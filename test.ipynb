```python
import os
import json
# read json files from a directory
data = []
path='./contracts-with-sources/'
for filename in os.listdir(path):
    with open(path + filename ,encoding='UTF-8') as f:
        print(len(data)+1, filename)
        try:
            data.append(json.load(f))
        except:
            print('Error reading file')
            continue

# write list as csv file
# with open('addresses.csv', 'w') as f:
#     f.write('\n'.join(data))
res=[]
for i in data:
    with open('./contracts-with-sources-extracted/'+i['ContractName']+'.sol','w') as f:
        try:
            f.write(i['SourceCode'])
        except:
            print('Error reading file')

        res+=[[i['ContractName'],i['SourceCode']]]

```

```python
# scrape data from etherscan
import requests
import bs4

base_address='https://ethervm.io/decompile/'
# read csv file
with open('addresses.csv', 'r') as f:
    addresses = f.read().split('\n')

# get data from ethervm
data = []
for address in addresses[:1]:
    url = base_address + address
    r = requests.get(url)
    soup = bs4.BeautifulSoup(r.text, 'html.parser')
    # get contract source code
    print(soup)
    source = soup.find('code')
    print(source)
```
